```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((100, 100, 3), dtype=np.uint8)
```

```python
plt.imshow(img)
```

```python
img[:30, :, 0] = 255

plt.imshow(img)
```

```python
img[30:60, :, 1] = 255

plt.imshow(img)
```

```python
img[60:, :, 2] = 255

plt.imshow(img)
```
